```python
def extract_moda(data: str) -> list:
  # Regex pattern to match the desired structure
  pattern = re.compile(r'(\w+)\r\n((?:\w+,?\s?)+)')

  # Find all matches
  matches = pattern.findall(data)

  # Convert matches to the desired format
  result = [(match[0], [x.strip() for x in match[1].split(',')]) for match in matches]

  return result
```

```python
df_moda = df.copy()
df_moda['antarmodas'] = df_moda['antarmodas'].apply(lambda x: extract_moda(str(x)))

df_moda = df_moda [['nid', 'title', 'antarmodas']]
df_moda
```

```python
# df_moda ['moda'] = df_moda['antarmodas'].apply(lambda x: [item[0] for item in x])
df_moda = df_moda.explode('antarmodas')
df_moda.dropna(inplace = True)
df_moda
```

```python
df_moda ['moda'] = df_moda['antarmodas'].apply(lambda x: x [0])

df_moda ['armada'] = df_moda['antarmodas'].apply(lambda x: x [1])

df_moda.drop(columns = ['antarmodas'], inplace = True)
df_moda.reset_index(drop = True, inplace = True)
df_moda
```

```python
df_moda.to_csv('Extract/moda.csv', index = False)
```
